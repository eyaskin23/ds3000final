## 3. Clean up data (preliminary): remove/imputate missing values

### Next we counted up the null values in each column of the dataframe. We did expect to see a lot of null values across the board since there were many columns and it wouldn't make sense for every column in most of the rows to be non-null. For example it is unlikely to see multiple types of overdose for the same entry in the dataframe. (i.e someone overdosed on Cocaine, and Heroin, and Fentanyl, etc.)

### The reason we wanted to do this is to determine which columns we wanted to clean up and fix, and which ones were not worth keeping for the project.

### We also used df['col_name'].value_counts() to help us determine if a column was useful

```python
# get the null value counts of each column
df.isnull().sum()
```

### `Date` and `DateType` are completely free of null values, and are very useful columns for trends so we decided we wanted both.

### `Age`, `Sex`, and `Race` have very few null values, and are very useful columns for demographics so we decided it was worth keeping and cleaning up. We fixed them up through imputation by setting missing values to the mode of each respective column.

```python
# Calculate the mode of age, sex, and race column
mode_age = df['Age'].mode()[0]
mode_sex = df['Sex'].mode()[0]
mode_race = df['Race'].mode()[0]

# Impute missing values with the mode
df['Age'].fillna(mode_age, inplace=True)

df['Sex'].fillna(mode_sex, inplace=True)

df['Race'].fillna(mode_race, inplace=True)
```

## 1. SVM Classification - `Polydrug use` as the target variable, creating the column

```python
columns_to_sum = df.columns[7:24]  # Adjust the indices as per the actual DataFrame

# Sum these columns for each row
drug_sum = df[columns_to_sum].sum(axis=1)

# Create a new column 'polydrug_use' which is True if the sum is 2 or more, else False
df['polydrug_use'] = drug_sum >= 2
```

- We will use age, sex, and race to see if we can predict polydrug use.
- We will take a look at the distribution of age as well
- We will split the data into train and test.
- Also making sure to encode any categorical labels and normalize any numerical labels.

### Prepping the data (selecting features, analyzing age, splitting/normalizing/encoding)

### `Note:` When trying to split data we had trouble encoding because some races had 1 value present (this meant that sometimes the train set would not know how to encode a `Korean` since it would be split into the test set and not the train set. To fix this we simply removed the bottom four (freqs. of 1) from our df. 

```python
df['Race'].value_counts()
```

```python
df = df[df['Race'] != 'American Indian or Alaska Native']
df = df[df['Race'] != 'Hawaiian']
df = df[df['Race'] != 'Native American, Other']
df = df[df['Race'] != 'Korean']
```

## `Note`: Our dataframe had an imbalanced label, meaning we had to remove (randomly) about 3/8 of our True samples

```python
df['polydrug_use'].value_counts()
```

```python
# Filter the DataFrame to get only the rows where 'polydrug_use' is True
true_samples = df[df['polydrug_use'] == True]

# Randomly select 5500 'True' samples to remove
samples_to_remove = true_samples.sample(n=5500, random_state=42)  # random_state for reproducibility

# Drop these samples from the original DataFrame
df_reduced = df.drop(samples_to_remove.index)

# Check the result
print(df_reduced['polydrug_use'].value_counts())
```

## We selected age, sex, and race for our features, ignoring location based parameters and categorizations like Cause of Death were too varied in response to be useful.

```python
# df_features = df[['Age', 'Sex', 'Race', 'Heroin', 'Cocaine', 'Fentanyl', 'Fentanyl Analogue', 'Oxycodone', 'Oxymorphone', 'Ethanol', 'Hydrocodone', 'Benzodiazepine', 'Methadone', 'Meth/Amphetamine', 'Amphet', 'Tramad', 'Hydromorphone', 'Morphine (Not Heroin)', 'Xylazine', 'Gabapentin', 'Opiate NOS']]
df_features = df_reduced[['Age', 'Sex', 'Race']]
```

```python
# Create a histogram
plt.hist(df_reduced['Age'], bins=30, edgecolor='black')
plt.title('Histogram of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

### Age appears to be roughly symmetric and multimodal. Most of the values are between 30 and 60, with a few outliers past 80 and at 12-14.

## Splitting into train/test

```python
#partition the data
X   = df_features #get the input features
y   = df_reduced['polydrug_use']              #get the target

X_train, X_test, y_train, y_test = train_test_split(X,              #the input features
                                                    y,              #the label
                                                    test_size=0.3,  #set aside 30% of the data as the test set
                                                    random_state=69, #reproduce the results
                                                    stratify=y      #preserve the distribution of the labels
                                                   )
```

## Encode categorical data and normalize numeric

```python
# Define which columns you want to one-hot encode and normalize
categorical_features = ['Sex', 'Race']  # columns to one-hot encode
numerical_features = ['Age']  # column to normalize

# Create a transformer for one-hot encoding categorical features
categorical_transformer = OneHotEncoder()

# Create a transformer for scaling numerical features
numerical_transformer = StandardScaler()

# Fit and transform the numerical features
X_train_numerical = numerical_transformer.fit_transform(X_train[numerical_features])
X_train_numerical_df = pd.DataFrame(X_train_numerical, columns=numerical_features, index=X_train.index)

# Fit and transform the categorical features
X_train_categorical = categorical_transformer.fit_transform(X_train[categorical_features]).toarray()
categorical_columns = categorical_transformer.get_feature_names_out(categorical_features)
X_train_categorical_df = pd.DataFrame(X_train_categorical, columns=categorical_columns, index=X_train.index)

# Combine the transformed numerical and categorical features into one DataFrame
X_train_transformed = pd.concat([X_train_numerical_df, X_train_categorical_df], axis=1)
```

## Apply encoder and scaler to test set

```python
# Transform the numerical features of X_test
X_test_numerical = numerical_transformer.transform(X_test[numerical_features])
X_test_numerical_df = pd.DataFrame(X_test_numerical, columns=numerical_features, index=X_test.index)

# Transform the categorical features of X_test
X_test_categorical = categorical_transformer.transform(X_test[categorical_features]).toarray()
categorical_columns = categorical_transformer.get_feature_names_out(categorical_features)
X_test_categorical_df = pd.DataFrame(X_test_categorical, columns=categorical_columns, index=X_test.index)

# Combine the transformed numerical and categorical features into one DataFrame for X_test
X_test_transformed = pd.concat([X_test_numerical_df, X_test_categorical_df], axis=1)
```

### View the prepared data

```python
#view the prepared data
X_train_transformed.head(10)
X_test_transformed.head(10)
```

## Training and evaluation

```python
svm = SVC() #the default kernel is rbf
svm.fit(X_train_transformed, y_train) #fit the data
```

```python
#predict the labels for the test set
y_pred = svm.predict(X_test_transformed)

print('The predicted Outcome is: {}'.format(y_pred))
```

```python
# evaluate the prediction accuracy
print('The accuracy of the model is: {}'.format(round(svm.score(X_test_transformed, y_test), 2)))
```

```python
#get the classification report
target_names = ["True", "False"]
print(classification_report(y_test, y_pred, target_names = target_names))
```

## Analysis of "default hyperparameters" model
Accuracy: The model's accuracy is 53%, which is the proportion of total correct predictions (both true positives and true negatives) out of all predictions made. This indicates that the model is only slightly better than random guessing, which would have an accuracy of 50% in a balanced dataset.

Precision: Precision measures the proportion of positive identifications that were actually correct. For the "True" class, the precision is 0.55, meaning that when the model predicts polydrug use, it is correct 55% of the time. For the "False" class, the precision is 0.52, indicating that the model is correct 52% of the time when predicting no polydrug use.

Recall (measures the proportion of actual positives that were identified correctly): For the "True" class, the recall is 0.38, which means that the model correctly identifies 38% of all actual polydrug use cases. For the "False" class, the recall is 0.68, indicating that it correctly identifies 68% of the no polydrug use cases.

F1-Score: The F1-score is the harmonic mean of precision and recall, providing a single metric to assess the balance between precision and recall. The F1-score for the "True" class is 0.45, and for the "False" class, it is 0.59. These scores suggest that the model is more balanced at predicting non-polydrug use cases but generally struggles to achieve a good balance between precision and recall, especially for the "True" class.

Support: The support is the number of actual occurrences of each class in the dataset. There are 781 cases of "True" and 764 cases of "False," indicating a relatively balanced dataset in terms of class distribution.

In summary, the initial untuned model shows limited effectiveness in predicting polydrug use. It performs slightly better at identifying non-polydrug use cases (False) than polydrug use cases (True). The relatively low recall for the "True" class suggests that many actual polydrug use cases are being missed, while the precision indicates that there are also a number of false positives in the predictions. Hopefully <b>tuning</b> of hyperparameters could enhance its predictive performance.

```python
# Define the parameter grid
param_grid = {
    'C': [0.1, 1, 10],  # values for C to test
    'gamma': [0.001, 0.1, 1, 10]  # values for gamma to test
}

svm = SVC()

#initialize Grid Search with Cross Validation and return the training score, k=5 fold
svm_grid = GridSearchCV(estimator = svm, param_grid = param_grid, cv=5, return_train_score = True) 
svm_grid.fit(X_train_transformed, y_train)

print(svm_grid.best_params_)
print(round(svm_grid.best_score_, 4))
```

```python
#display the cv results for the train and test set
cv_results = pd.DataFrame(svm_grid.cv_results_)
cv_results = cv_results.sort_values('mean_test_score', ascending=False)
cv_results[['mean_train_score', 'std_train_score', 'mean_test_score', 'std_test_score']].head(10)
```

```python
y_pred = svm_grid.predict(X_test_transformed)
print(y_pred)
print(classification_report(y_test, y_pred, target_names = target_names))
```

### Analysis of tuned model
The values in the classification report indicate that the SVM model has a moderate level of performance in predicting polydrug use, with a slight lean towards better performance in identifying non-polydrug users (False class) over polydrug users (True class).

The precision for the True class is higher than the False class, meaning the model is relatively better at ensuring that when it predicts polydrug use, it is correct. However, its recall for the True class is quite low, suggesting the model misses a significant number of actual polydrug use cases, labeling them incorrectly as non-polydrug use. This is a concern as it indicates a substantial rate of false negatives in the context of polydrug use prediction.

Conversely, the model has a better recall for the False class than the True class, indicating it is more effective at identifying non-polydrug users. However, its precision for the False class is only slightly above 50%, which means there is a considerable number of non-polydrug cases that are incorrectly predicted as polydrug use (false positives).

The overall accuracy of 53% is only slightly better than a random guess, and the macro and weighted averages for precision, recall, and F1-score are around 53%, indicating a balanced but moderate performance across both classes.

In the bigger picture, these metrics suggest that while the model has some ability to distinguish between polydrug and non-polydrug use, it is not highly reliable. It tends to miss more actual cases of polydrug use but is better at catching non-use cases. For practical applications, especially in healthcare or law enforcement, where missing cases of polydrug use could have serious implications, the model’s performance may not be satisfactory. Improvements in the model are necessary to ensure it can more accurately identify polydrug use without a significant number of false negatives or positives.

```python
y_pred = svm_grid.predict(X_train_transformed)
print(y_pred)
print(classification_report(y_train, y_pred, target_names = target_names))
```

## 2. KNN Classification - `Polydrug use` as the target variable

```python
# Define the model
knn = KNeighborsClassifier()

# Define the parameter grid: here 'n_neighbors' is the parameter we want to tune
param_grid = {'n_neighbors': range(3, 50, 2)}

# Create the GridSearchCV object
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')  # cv=5 for 5-fold cross-validation

# Fit the model using GridSearchCV
grid_search.fit(X_train_transformed, y_train)

# Get the best parameter
best_k = grid_search.best_params_['n_neighbors']
print('The best value for k is:', best_k)

# Get the model with the best parameter
best_model = grid_search.best_estimator_

# Evaluate the model
print('The accuracy of the model for k = {} is: {}'.format(best_k, best_model.score(X_test_transformed, y_test)))
```

### 3. Decision Tree Classification - `Polydrug use` as the target variable

```python
# Initialize your Decision Tree Classifier
dt = DecisionTreeClassifier(random_state=7)

# Define the parameter grid to search over
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(dt, param_grid, cv=5, scoring='accuracy')

# Fit GridSearchCV on the training data
grid_search.fit(X_train_transformed, y_train)

# Get the best estimator
best_dt = grid_search.best_estimator_

print(best_dt)

# Predict the labels for the test set using the best model found
y_pred = best_dt.predict(X_test_transformed)
print('Polydrug prediction: {}'.format(y_pred))

# Generate the classification report
print(classification_report(y_test, y_pred, target_names=target_names))
```
